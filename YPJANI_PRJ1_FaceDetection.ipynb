```python
# to solve this error we add 'minNeighbors'

image = cv2.imread('people2.jpg')

image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
detections = face_detector.detectMultiScale(image_gray, scaleFactor = 1.2,minNeighbors=7)
display (detections)
for (x, y, w, h) in detections:
    print(w, h)
    cv2.rectangle(image, (x, y), (x + w, y + h), (0,255,0), 2)
cv2.imshow('',image)
cv2.waitKey(10000)
plt.show()
cv2.destroyAllWindows()

# But now also some faces aren't detected
```

```python
# to solve this error we add 'minSize' & 'maxSize'

image = cv2.imread('people2.jpg')
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
detections = face_detector.detectMultiScale(image_gray, scaleFactor = 1.2,
                                            minNeighbors=2,minSize=(40,40), maxSize=(100,100))
display (detections)
for (x, y, w, h) in detections:
    print(w, h)
    cv2.rectangle(image, (x, y), (x + w, y + h), (0,255,0), 2)
cv2.imshow('',image)
cv2.waitKey(30000)
plt.show()
cv2.destroyAllWindows()

# we try to detect all faces but 1 face is not detected yet, so we need more Advaced Detection tool which we see further.
```

```python
#                                             Face Detection Model 02 with HOG and dlib         
```

```python
# display the original image

image = cv2.imread('people2.jpg')
cv2.imshow('',image)
cv2.waitKey(10000)
plt.show()
cv2.destroyAllWindows()
```

```python
# No. of Detected Faces and its Coordinates

face_detector_hog = dlib.get_frontal_face_detector()
detections = face_detector_hog(image, 1) # 1  is the scale factor 
print("No of Faces Detected in image 'people 2' :")
display (len(detections))
print()
display (detections)
```

```python
# display All detected Face with rectangle

for face in detections:
  l, t, r, b = face.left(), face.top(), face.right(), face.bottom()
  cv2.rectangle(image, (l, t), (r, b), (0, 255, 255), 2)
cv2.imshow('',image)
cv2.waitKey(10000)
plt.show()
cv2.destroyAllWindows()
```

```python
#                                               Face Detection Model 02 with CNN and dlib
```

```python
# display the original image

image = cv2.imread('people2.jpg')
cv2.imshow('',image)
cv2.waitKey(10000)
plt.show()
cv2.destroyAllWindows()
```

```python
# load face detector (mmod_human_face_detector) .dat file

cnn_detector = dlib.cnn_face_detection_model_v1('D:\\01 YPJANI _ Educational\\Acmgrade\\Project\\PRJ Face Detection\\Weights\\mmod_human_face_detector.dat')
display (cnn_detector)
```

```python
# display All detected Face with rectangle

detections = cnn_detector(image, 1)
for face in detections:
  l, t, r, b, c = face.rect.left(), face.rect.top(), face.rect.right(), face.rect.bottom(), face.confidence
  print(c)
  cv2.rectangle(image, (l, t), (r, b), (255,255,0), 2)
cv2.imshow('',image)
cv2.waitKey(10000)
plt.show()
cv2.destroyAllWindows()
```
